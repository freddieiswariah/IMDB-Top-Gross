# Web Scraper for the top 50 highest grossing movies on IMDB

```python
from bs4 import BeautifulSoup
import urllib.request
import csv
import pandas as pd
import re
```

```python
urlpage =  'https://www.imdb.com/search/title/?title_type=feature&genres=action&sort=boxoffice_gross_us,desc&explore=genres'
```

```python
# query the website and return the html to the variable 'page'
page = urllib.request.urlopen(urlpage)
# parse the html using beautiful soup and store in variable 'soup'
soup = BeautifulSoup(page, 'html.parser')
```

```python
print(soup.prettify())
```

```python
movies = soup.find_all('div', attrs={'class': 'lister-item-content'})
movies[0]
```

```python
# Initialise some empty lists 
titles=[]
years=[]
certs=[]
run_times=[]
genres=[]
imdb_scores=[]
metascores=[]
descriptions=[]
directors=[]
stars=[]
votes=[]
gross=[]

# A big for loop to extract data for each movie

for movie in movies:
    # Get the titles of the movies
    name = movie.find('h3', attrs={'class': 'lister-item-header'})
    title = name.find('a').string
    titles.append(title)
    # Get the year
    year = name.find('span', attrs={'class': 'lister-item-year text-muted unbold'}).string
    years.append(year)
    # Get the certificate
    info = movie.find('p', attrs={'class': 'text-muted'})
    cert = info.find('span', attrs={'class': 'certificate'}).string
    certs.append(cert)
    # Get the run time
    run_time = info.find('span', attrs={'class': 'runtime'}).string
    run_time = run_time.strip(' min')
    run_times.append(run_time)
    # Get the genres (and tidy up a bit)
    genre = info.find('span', attrs={'class': 'genre'}).string
    genres.append(genre.replace('\n', '').replace('            ', ''))
    # Get the ratings (again tidy up by removing empty spaces)
    imdb_score = movie.find('strong').string
    imdb_scores.append(imdb_score)
    score = movie.find('div', attrs={'class': 'inline-block ratings-metascore'})
    metascore = score.find('span').string
    metascores.append(metascore.replace('        ', ''))
    # Get the description
    description = movie.find_all('p', attrs={'class': 'text-muted'})
    description = description[1]
    description = description.text
    descriptions.append(description)
    # Get the director
    people = movie.find('p', attrs={'class': ''})
    director = people.find('a').string
    directors.append(director)
    # Get the vote count
    data_all = movie.find_all('span', attrs={'name': 'nv'})
    data1 = data_all[0]
    data1 = data1.string.replace(',', '')
    votes.append(data1)
    # Get the gross revenue
    data2 = data_all[1].string
    gross.append(data2)
    rows.append([titles, years, certs, run_times, genres, imdb_scores, metascores, descriptions, directors, votes, gross])
```
